# Import Libraries

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

# Getting Started With Machine Learning 

```python
# define the input data and output data
xs = np.array([-1, 0, 1,2,3,4], dtype=float)
ys = np.array([-3,-1,1,3,5,7], dtype=float)
```

Relasi dari X dan Y adalah Y = 2X - 1

```python
# define the model architecture
model = Sequential([
    Dense(units=1, input_shape=[1])
])
model.compile(
    optimizer='sgd',
    loss='mean_squared_error',
)

# fitting data into model
model.fit(xs, ys, epochs=100)
```

```python
# print prediction of new data
print(model.predict([10]))
```

**Explanation [Indonesia]** <br>
1. Ketika menggunakan Tensorflow, kita mendefinisikan layer-layer dengan **Sequential**
2. Model diatas hanya memiliki 1 layer (Dense) dalam Sequential artinya arsitektur model hanya memiliki 1 layer.
3. Dengan *keras.layers.API* kita bisa mendefinisikan tipe layer yang akan dibuat (Convolutional, Recurent, dll). Kali ini dibuat tipe **Dense** layer yang artinya sekumpulan neuron-neuron yang saling terhubung satu sama lainnya.
4. Dense layer yang dibuat memiliki **units=1**, artinya hanya ada 1 neuron di layer tersebut.
5. Kesimpulannya, arsitektur modelnya adalah 1 dense layer dengan 1 neuron keseluruhan.

# Seeing What the Network Learned
***
Hubungan yang bisa dilihat dari X dan Y adalah Y = 2X - 1. Dalam hal ini, neural network memiliki parameter weight dan bias yang akan dipelajari untuk bisa mendapatkan weight mendekati 2 dan bias mendekati -1. <br>
Dengan menggunakan tensorflow, bisa dilihat weight dan bias yang telah dipelajari.

```python
# define the model architecture
l0 = Dense(units=1, input_shape=[1])
model2 = Sequential([
    l0
])
model2.compile(
    optimizer='sgd',
    loss='mean_squared_error',
)

# fitting data into model
model2.fit(xs, ys, epochs=100)
```

```python
# print prediction of new data
print(model2.predict([10]))

# print the weight and bias they are learned
print(l0.get_weights())
```

Learned_weight = 1,8 and Learned_bias = -0,5. Almost 2 and -1!
